```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([5,4,9,8,11])
b = np.array(10*a[::-1])

print(a)
print(b)

c = np.corrcoef(a,b)
print("")
print(c,"\n")

plt.figure(figsize=(5, 5))
plt.plot(c)
plt.title("Correlation Graph")
plt.show()
```

```python
x_pos = [1, 2, 3, 4, 5, 6, 7]
y_pos = [i+1 for i in x_pos]

x_neg = [1, 2, 3, 4, 5, 6, 7]
y_neg = [x_neg[::-1]]

x_zero = np.array(np.random.randint((10),size=(20)))
y_zero = np.array(np.random.randint((10),size=(20)))
```

```python
# Plotting

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.scatter(x_pos, y_pos, color='blue')
plt.title("Positive Correlation")

plt.subplot(1, 3, 2)
plt.scatter(x_neg, y_neg, color='red')
plt.title("Negative Correlation")

plt.subplot(1, 3, 3)
plt.scatter(x_zero, y_zero, color='green')
plt.title("Zero Correlation")

plt.tight_layout()
plt.show()
```

```python
arr1 = np.array(np.random.randint((1000), size=(500)))
arr2 = np.array(np.random.choice((arr1*57), size=(500)))

plt.scatter(arr1,arr2)
```

```python
rel = np.corrcoef(arr1,arr2)
print(rel)
plt.plot(rel)
```

```python
import pandas as pd
df=pd.DataFrame({'a':np.random.randint(0,50,1000)})
df['b']=df['a']+np.random.randint(0,50,1000)
df['c']=100-df['a']+np.random.randint(0,5,1000)
df['d']=np.random.randint(0,50,1000)
df.corr()
```

```python
pd.plotting.scatter_matrix(df,figsize=(6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
plt.show()
```
